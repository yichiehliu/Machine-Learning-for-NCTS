```python
#酒的酒精濃度、顏色、色彩強度的回歸分析

from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression


wine = datasets.load_wine()
df1 = pd.DataFrame(wine['data'],columns=wine['feature_names'])

#資料顯示
df1
```

```python
#回歸分析
x = df1[['color_intensity','hue']]
y = df1['alcohol']
lm = LinearRegression()
model=lm.fit(x, y)

# 印出係數
print('係數',lm.coef_[0], lm.coef_[1])

# 印出截距
print('截距',lm.intercept_ )

pd.DataFrame(lm.predict(x), columns = ['prediction'])

#3D資料視覺化
from mpl_toolkits.mplot3d import Axes3D
x1 = np.array(df1['color_intensity']) 
x2 = np.array(df1['hue'])
y = np.array(df1['alcohol'])
td = plt.subplot(projection='3d')  
td.scatter(x1, x2, y)
```
